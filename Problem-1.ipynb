```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1D Random walk

We will concider two kinds of 1D random talk.
* $dx$ is uniformly distributed between $[-1/2, +1/2]$.
* $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2.

For both kinds of the 1D random walk, you should do the following:
* Plot a single trajectory.
* Plot the histogram of the end point for an ensemble of trajectories.

```python
def random_1D(steps_per_run):
    x_step = np.random.random(steps_per_run)-1/2
    return(np.cumsum(x_step))

steps_per_run = 1000
number_of_runs = 10000

x_stop = np.zeros(number_of_runs)
for i in range(number_of_runs):
    x = random_1D(steps_per_run)
    x_stop[i] = x[-1]
    
fix, ax = plt.subplots(figsize=(8,8))

ax.plot(random_1D(steps_per_run), '.')
plt.title(r'A single trajectory within [$-\frac{1}{2}, \frac{1}{2}$]', fontsize=16)

fix, ay = plt.subplots(figsize = (8,8))
_ = ay.hist(x_stop, bins=50)
plt.title('Histogram of all endpoints', fontsize=16)
```

```python
steps_per_run = 1000
number_of_runs = 10000
step_set = np.array([-1,1])

def random_1D(steps_per_run):
    x_step = np.random.choice(a = step_set, size = steps_per_run)
    return(np.cumsum(x_step))


x_stop = np.zeros(number_of_runs)
for i in range(number_of_runs):
    x = random_1D(steps_per_run)
    x_stop[i] = x[-1]
    
fiz, ax = plt.subplots(figsize=(8,8))

ax.plot(random_1D(steps_per_run), '.')
plt.title(r'A single trajectory with step = -1 or 1', fontsize=16)

fix, ay = plt.subplots(figsize = (8,8))
_ = ay.hist(x_stop, bins=50)
plt.title('Histogram of all endpoints', fontsize=16)
```
