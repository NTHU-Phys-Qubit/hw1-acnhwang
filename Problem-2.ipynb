# 2D Random walk

We will concider three kinds of 2D random talk.
* $dx$ and $dy$ are uniformly distributed between $[-1/2, +1/2]$.
* $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2. Similarly for $dy$.
* $(dx, dy)=(\cos(\theta), \sin(\theta))$ where $\theta$ is uniformly distributed between $[0, 2\pi]$.

For all kinds of the 2D random walk, you should do the following:
* Plot a single trajectory.
* Plot the distribution of the end point for an ensemble of trajectories.
* Plot the histogram of the $r=x^2_f+y^2_f$ for an ensemble of trajectories.

```python
def random_2D(steps_per_run):
    x_step = np.random.random(steps_per_run)-1/2
    y_step = np.random.random(steps_per_run)-1/2
    return(np.cumsum(x_step), np.cumsum(y_step))

steps_per_run = 1000
number_of_runs = 10000

x_stop = np.zeros(number_of_runs)
y_stop = np.zeros(number_of_runs)
r_stop = np.zeros(number_of_runs)
for i in range(number_of_runs):
    x, y = random_2D(steps_per_run)
    x_stop[i], y_stop[i] = x[-1], y[-1]
    r_stop[i] = np.sqrt(x_stop[i]**2 + y_stop[i]**2)

fix, ax = plt.subplots(figsize=(8,8))
ax.plot(random_2D(steps_per_run)[0], random_2D(steps_per_run)[1], '.')
plt.title(r'A single trajectory with dx, dy$\in [-\frac{1}{2}, \frac{1}{2}$]', fontsize=16)

fix, ay = plt.subplots(figsize=(8,8))
ay.plot(x_stop, y_stop, '.')
plt.title('Distribution of all end points', fontsize=16)

fix, az = plt.subplots(figsize = (8,8))
az.hist(r_stop, bins=50)
plt.title(r'Histogram of all endpoints of $r=\sqrt{x_f^2 + y_f^2}$', fontsize=16)
```

```python
step_set = np.array([-1, 1])

def random_2D(steps_per_run):
    x_step = np.random.choice(a = step_set, size = steps_per_run)
    y_step = np.random.choice(a = step_set, size = steps_per_run)
    return(np.cumsum(x_step), np.cumsum(y_step))

steps_per_run = 1000
number_of_runs = 10000

x_stop = np.zeros(number_of_runs)
y_stop = np.zeros(number_of_runs)
r_stop = np.zeros(number_of_runs)
for i in range(number_of_runs):
    x, y = random_2D(steps_per_run)
    x_stop[i], y_stop[i] = x[-1], y[-1]
    r_stop[i] = np.sqrt(x_stop[i]**2 + y_stop[i]**2)

fix, ax = plt.subplots(figsize=(8,8))
ax.plot(random_2D(steps_per_run)[0], random_2D(steps_per_run)[1], '.')
plt.title(r'A single trajectory with dx, dy$\in\pm1$ and probability=$\pm\frac{1}{2}$ respectively', fontsize=16)

fix, ay = plt.subplots(figsize=(8,8))
ay.plot(x_stop, y_stop, '.')
plt.title('Distribution of all end points', fontsize=16)

fix, az = plt.subplots(figsize = (8,8))
az.hist(r_stop, bins=50)
plt.title(r'Histogram of all endpoints of $r=\sqrt{x_f^2 + y_f^2}$', fontsize=16)
```

```python
def random_2D(steps_per_run):
    steps = np.random.random(steps_per_run) * 2 * math.pi
    x_step = np.cos(steps)
    y_step = np.sin(steps)
    return(np.cumsum(x_step), np.cumsum(y_step))

steps_per_run = 1000
number_of_runs = 10000

x_stop = np.zeros(number_of_runs)
y_stop = np.zeros(number_of_runs)
r_stop = np.zeros(number_of_runs)
for i in range(number_of_runs):
    x, y = random_2D(steps_per_run)
    x_stop[i], y_stop[i] = x[-1], y[-1]
    r_stop[i] = np.sqrt(x_stop[i]**2 + y_stop[i]**2)

fix, ax = plt.subplots(figsize=(8,8))
ax.plot(random_2D(steps_per_run)[0], random_2D(steps_per_run)[1], '.')
plt.title(r'A single trajectory with (dx,dy)=($cos\theta,sin\theta$) where $\theta\in [0, 2\pi$]', fontsize=16)

fix, ay = plt.subplots(figsize=(8,8))
ay.plot(x_stop, y_stop, '.')
plt.title('Distribution of all end points', fontsize=16)

fix, az = plt.subplots(figsize = (8,8))
az.hist(r_stop, bins=50)
plt.title(r'Histogram of all endpoints of $r=\sqrt{x_f^2 + y_f^2}$', fontsize=16)
```
